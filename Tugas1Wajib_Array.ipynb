```python
#Masukkan 10 Nilai
nilai_mahasiswa = []
for i in range(10):
  n = int(input(f"masukkan nilai mahasiswa ke-{i+1} :"))
  nilai_mahasiswa.append(n)
```

```python
#Tampilkan Nilai Tertinggi dan Terendah
print("nilai_mahasiswa :", nilai_mahasiswa)
print("nilai maksimal :", max(nilai_mahasiswa))
print("nilai minimal :", min(nilai_mahasiswa))
```

```python
#Menghitung Rata-rata
print("rata-rata :", sum(nilai_mahasiswa) / len(nilai_mahasiswa))
```

```python
#Menghitung Jumlah Siswa Lulus (>= 60)
lulus = 0
tidak_lulus = 0

for n in nilai_mahasiswa:
    if n >= 60:
        lulus += 1
    else:
        tidak_lulus += 1
print("jumlah lulus :", lulus)
print("jumlah tidak_lulus :", tidak_lulus)
```

```python
#Grafik untuk menggambarkan nilai tertinggi dan terendah
import matplotlib.pyplot as plt

plt.figure()
kategori = ["Tertinggi", "Terendah"]
nilai_tertinggi = max(nilai_mahasiswa)
nilai_terendah = min(nilai_mahasiswa)
data = [nilai_tertinggi, nilai_terendah]

plt.bar(kategori, data)
plt.title("Grafik Nilai Tertinggi dan Terendah")
plt.ylabel("Nilai")
plt.ylim(0, 100)
plt.show()
```

```python
#Buat grafik untuk menggambarkan data kelulusan
import matplotlib.pyplot as plt

plt.figure()
kategori = ["Lulus", "Tidak_Lulus"]
data = [lulus, tidak_lulus]

plt.bar(kategori, data)
plt.title("Grafik Data Kelulusan Mahasiswa")
plt.ylabel("Jumlah Mahasiswa")
plt.ylim(0, 10)
plt.show()
```

```python
#Analisis Kompleksitas
print("\nAnalisis Kompleksitas:")
print("- Input 10 nilai: O(10) ~ O(1) karena jumlah tetap")
print("- Menampilkan nilai Tertinggi dan Terendan: O(n)")
print("- Menghitung rata-rata: O(n)")
print("- Menghitung jumlah mahasiswa lulus: O(n)")
print("- Membuat grafik: O(1) untuk data kecil (10 elemen)")
```
